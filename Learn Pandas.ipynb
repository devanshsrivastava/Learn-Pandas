```python
#Creating New DataFrame

india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
#Now we will Concatenate both DataFrames

df2 = pd.concat([india_weather, us_weather])
df2
```

```python
#Note above the Index is not continuous, so to get continuous index
df2 = pd.concat([india_weather, us_weather], ignore_index= True)
df2
```

```python
df2.loc[0]
```

```python
import pandas as pd
df_merge1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df_merge1
```

```python
df_merge2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df_merge2
```

```python
#We can merge two dataframes to get more than for one columns for one main column like city over here

df_merge = pd.merge(df_merge1,df_merge2,on = 'city')
df_merge
```

```python
#You can also use concept of joins like inner and outer to merge fields
#Also can use right and left join concepts

df_merge = pd.merge(df_merge1,df_merge2,on = 'city',how = 'inner')
df_merge
```

```python
df_survey = pd.read_excel("survey.xls")
df_survey
```

```python
#Learning CrossTabs

pd.crosstab(df_survey.Nationality,df_survey.Handedness)
```

```python
pd.crosstab(df_survey.Sex,df_survey.Handedness)
pd.crosstab([df_survey.Nationality, df_survey.Sex], [df_survey.Handedness], margins=True)
```

```python
pd.crosstab(df_survey.Sex, df_survey.Handedness, normalize='index')
```

```python
#Aggfunc and Values

import numpy as np
pd.crosstab(df_survey.Sex, df_survey.Handedness, values=df_survey.Age, aggfunc=np.average)
```
